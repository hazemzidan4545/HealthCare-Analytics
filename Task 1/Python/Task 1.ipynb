```python
import pandas as pd
import requests
from faker import Faker
import random
from sqlalchemy import create_engine
import time
```

```python
df = pd.read_csv(r'KaggleV2-May-2016.csv')
```

```python
df.head(20)
```

```python
df.shape
```

```python
fake = Faker()
#create fake specs for doctors
specialties = ["Cardiologist", "Dermatologist", "General Practitioner", 
               "Neurologist", "Orthopedic", "Pediatrician"]
#variables to hold el names, specs, hospitals
doctor_names = []
doctor_specialties = []
doctor_hospitals = []

#generate fake data
for i in range(len(df)):
    name = fake.name()
    specialty = random.choice(specialties)
    hospital = fake.company() + " Hospital"
    
    doctor_names.append(name)
    doctor_specialties.append(specialty)
    doctor_hospitals.append(hospital)
    
#inject fake data into dataframe
df['DoctorName'] = doctor_names
df['Specialty'] = doctor_specialties
df['Hospital'] = doctor_hospitals
```

```python
#dataframe has no null values, but we'll assume positive
df = df.dropna()
```

```python
#converting these columns datatypes to boolean to reserve resources bc they carry 0s, and 1s though assigned as int64 
binary_columns = ['Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']

#convert to boolean
df[binary_columns] = df[binary_columns].astype(bool)
```

```python
df.dtypes
```

```python
#age group categorization
bins = [0, 12, 19, 30, 60, float('inf')]
labels = ['0-12', '13-19', '20-30', '31-60', '60+']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
```

```python
df
```

```python
#moving age catorization next to the "Age" column from the most right
df.insert(df.columns.get_loc('Age') + 1, 'AgeGroup', df.pop('AgeGroup'))
```

```python
#converting IDs into int64 for easier querying and insights
df['PatientId'] = df['PatientId'].astype('int64')
pd.set_option('display.float_format', '{:,.0f}'.format)
```
